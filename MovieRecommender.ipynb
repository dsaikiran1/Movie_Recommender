```python
import ast

def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies_data['genres'].apply(convert)
```

```python
movies_data['genres'] = movies_data['genres'].apply(convert)
```

```python
import ast
def convert_cast(text):
    counter = 0
    l=[]
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter+=1
    return l
```

```python
movies_data['cast'] = movies_data['cast'].apply(convert_cast)
movies_data['cast']
```

```python
movies_data['crew'].iloc[0]
```

```python
movies_data['keywords'] = movies_data['keywords'].apply(convert)
movies_data['keywords']
```

```python
import ast

def convert_crew(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l

        
```

```python
movies_data['crew']=movies_data['crew'].apply(convert_crew)
```

```python
movies_data.dropna(inplace=True)
```

```python
movies_data['overview'] = movies_data['overview'].apply(lambda x:x.split())
```

```python
movies_data['overview']
```

```python
def remove_space(text):
    l=[]
    for i in text:
        l.append(i.replace(" ",""))
    return l
    
```

```python
movies_data['cast'] = movies_data['cast'].apply(remove_space)
movies_data['crew'] = movies_data['crew'].apply(remove_space)
movies_data['genres'] = movies_data['genres'].apply(remove_space)
movies_data['keywords'] = movies_data['keywords'].apply(remove_space)
```

```python
movies_data.head()
```

```python
movies_data['tags'] = movies_data['overview']+movies_data['genres']+movies_data['keywords']+movies_data['cast']+movies_data['crew']
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0].tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```
